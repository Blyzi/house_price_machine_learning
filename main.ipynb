### Data preparation

```python
# Change the type of the columns

df['Date mutation'] = pd.to_datetime(df['Date mutation'], format='%d/%m/%Y')
df['Code postal'] = df['Code postal'].astype(str)
df['Code departement'] = df['Code departement'].astype(str)
df['Code commune'] = df['Code commune'].astype(str)
df['Valeur fonciere'] = df['Valeur fonciere'].replace(',', '.', regex=True).astype(float)

df.head(10)
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
# Delete duplicate rows

df.drop_duplicates(inplace=True)
df.duplicated().sum()
```

```python
# Code INSEE

df['Code INSEE'] = df.apply(lambda x: x['Code departement'].rjust(2, '0') + x['Code commune'].rjust(3, '0'), axis=1)
```

```python
# Use geopandas to get the coordinates of the communes

shapefile = 'communes/communes-20220101.shp'
gdf = gpd.read_file(shapefile)[['insee', 'geometry']]

gdf = gdf.merge(df, left_on='insee', right_on='Code INSEE', how='inner')

gdf.head(10)

```
